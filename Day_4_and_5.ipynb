```python
#Cleaning of review data set
import pandas as pd
df = pd.read_csv('user_reviews.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Dropping empty rows because if all rows which are empty are consistent in the columns
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#putting all the data in "Sentiment_Polarity" and "Sentiment_Subjectivity" columns to 3 decimal places for consistency in data
df['Sentiment_Polarity'] = df['Sentiment_Polarity'].round(3)
df['Sentiment_Subjectivity'] = df['Sentiment_Subjectivity'].round(3)
```

```python
df.head(20)
```

```python
#Cleaning of apps dataset
df2 = pd.read_csv('apps.csv')
```

```python
df2.info()
```

```python
#Removing the unnamed column
df2.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#Interpolating missing values in ratings and size column
df2['Rating'] = df2['Rating'].interpolate(method='linear')
df2['Size'] = df2['Size'].interpolate(method='linear')
```

```python
df2.isnull().sum()
```

```python
#rounding up size column data because the size of an app is a whole number
import math
df2['Size'] = df2['Size'].apply(lambda x: math.ceil(x))
df2.head(10)
```

```python
#Standardizing data in the Last Updated column
df2['Last Updated'] = pd.to_datetime(df2['Last Updated'])
df2['Last Updated'] = df2['Last Updated'].dt.strftime('%Y-%m-%d')
df2.head(10)
```

```python
df2.dropna(inplace=True)
```

```python
df2['Android Ver'] = df2['Android Ver'].str.replace('and up', '')
df2.head()
```

```python
#Standardizing data in the current version and android version columns
def format_version(version):
    if version == 'Varies with device':
        return version
    else:
        parts = version.split('.')
        while len(parts) < 3:
            parts.append('0')
        return '.'.join(parts[:3])

df2['Current Ver'] = df2['Current Ver'].apply(format_version)
df2['Android Ver'] = df2['Android Ver'].apply(format_version)
df2.head(10)
```
